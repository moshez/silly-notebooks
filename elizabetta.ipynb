```python
from PIL import Image
```

```python
img = Image.open("data/mona-lisa.jpg")
```

```python
x, y = img.size
```

```python
img.resize((x//10, y//10))
```

```python
eyes = Image.open("data/eyes.png")
eyes = eyes.convert("RGB")
eye_x, eye_y = eyes.size
eye_x, eye_y
```

```python
eyes
```

```python
with_eyes = img.convert("RGB")
with_eyes.paste(eyes.resize((int(eye_x // 1.5), int(eye_y // 1.5))), 
                ((525, 420)))
with_eyes.resize((x//5, y//5))
```

```python
with_eyes.save("mona-lisa-with-eyes.png")
```
